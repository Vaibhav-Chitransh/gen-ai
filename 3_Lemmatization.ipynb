```python
# Wordnet Lemmatizer
# This techinque is like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem.
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
# POS: noun - n, verb - v, adjective - a, adverb - r (by default pos = "n")
lemmatizer.lemmatize("going", pos="v")
```

```python
lemmatizer.lemmatize("going", pos="n")
```

```python
words = ["programming", "programs", "history", "finally", "finalize", "writing", "writes", "eating", "eats", "eaten"]
for word in words:
    print(word + "-->" + lemmatizer.lemmatize(word, pos="v"))
print("\n")
for word in words:
    print(word + "-->" + lemmatizer.lemmatize(word, pos="n"))
```
