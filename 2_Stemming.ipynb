```python
# Classification problem
# classify the comments on products whether positive or negative
# stemming -> reducing the word to it's word stem
# [going, goes, gone] -> go,  [eating, eaten, eat] -> eat
words = ["programming", "programs", "history", "finally", "finalize", "writing", "writes", "eating", "eats", "eaten"]
```

```python
# PorterStemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + "-->" + stemming.stem(word))
```

```python
# some words like history -> histori (entire meaning got changed which is the major disadvantage of stemming)
stemming.stem("congratulations")
```

```python
# Regexp Stemmer -> basically takes a single regular expression and removes any prefix or suffix that matches the expression
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")   # $ is at the end so ing will be removed if it is at the end
```

```python
# Snowball Stemmer -> better than Porter Stemmer
from nltk.stem import SnowballStemmer
snowballStemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(word + "-->" + snowballStemmer.stem(word))
```

```python
# now you might say what's the difference even here also history -> histori then how is it better than PorterStemmer
# below you see the difference
```

```python
# Porter Stemmer
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
# Snowball Stemmer
snowballStemmer.stem("fairly"), snowballStemmer.stem("sportingly")
```

```python
# but snowball Stemmer also is not perfect as you see with history also
snowballStemmer.stem("goes")   # not perfect (that's why we use Lemmatization)
```
