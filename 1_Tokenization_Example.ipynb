```python
corpus = """Hello Welcome, to Vaibhav Chitransh's github repo.
Please do read the entire code! to learn tokenization.
"""
```

```python
print(corpus)
```

```python
# Tokenization
# Paragraph -> Sentence
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
print(documents)
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Tokenization
# Paragraph -> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
# sentence -> words
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)    # now 's will also get splitted into ["'", "s"]
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)   # . (full stop) will not be considered as a separate word except the last word
```
